# Library Book Management Using Tkinter and Database

```python
from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
```

```python
root = Tk()
root.title('A Library Applicant')
```

```python
def Database():
    global conn, cursor
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS books (library_id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, Book_Id TEXT, BookTitle TEXT, Author TEXT, Status TEXT, ISBN TEXT, Year TEXT)")
    
def Add():
    if BOOK_ID.get() == "" or BOOKTITLE.get() == "" or AUTHOR.get() == "" or STATUS.get() == "" or ISBN.get() == "" or YEAR.get() == "":
        txt_result.config(text="Please enter all the fields" ,fg="red")
    else:
        Database()
        cursor.execute("INSERT INTO books (book_id, booktitle, author, status, ISBN, year) VALUES (?,?,?,?,?,?)",(str(BOOK_ID.get()),str(BOOKTITLE.get()),str(AUTHOR.get()),str(STATUS.get()),str(ISBN.get()),str(YEAR.get())))
        conn.commit()
        BOOK_ID.set("")
        BOOKTITLE.set("")
        AUTHOR.set("")
        STATUS.set("")
        ISBN.set("")
        YEAR.set("")
        cursor.close()
        conn.close()
        txt_result.config(text="Record Inserted !",fg="green")

            
def Display():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM books ORDER BY author ASC")
    fetch = cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[1],data[2],data[3],data[4],data[5],data[6]))
    cursor.close()
    txt_result.config(text="Successfully fetched books !",fg="black")

    
def Issue():
    tree.delete(*tree.get_children())
    if YEAR.get()=="":
        txt_result.config(text="Please dont leave YEAR blank.",fg="red")
    else:
        Database()
        cursor.execute("SELECT * FROM books WHERE year=?",(str(YEAR.get()),))
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert('',"end",values=(data[1],data[2],data[3],data[4],data[5],data[6]))
        cursor.close()
        BOOK_ID.set("")
        BOOKTITLE.set("")
        AUTHOR.set("")
        STATUS.set("")
        YEAR.set("")
        ISBN.set("")
        txt_result.config(text="Successfully fetched data.",fg="black")
    
def Search():
    tree.delete(*tree.get_children())
    if BOOKTITLE.get()=="" or AUTHOR.get()=="" or YEAR.get()=="":
        txt_result.config(text="Please dont leave blank.",fg="red")
    else:
        if ISBN.get()=="":
            Database()
            cursor.execute("SELECT * FROM books WHERE booktitle=? AND author=? AND year=?",(str(BOOKTITLE.get()),str(AUTHOR.get()),str(YEAR.get())))
            fetch=cursor.fetchall()
            for data in fetch:
                tree.insert('',"end",values=(data[1],data[2],data[3],data[4],data[5],data[6]))
            cursor.close()
            BOOK_ID.set("")
            BOOKTITLE.set("")
            AUTHOR.set("")
            STATUS.set("")
            YEAR.set("")
            ISBN.set("")
            txt_result.config(text="Successfully fetched data.",fg="black")
        else :    
            Database()
            cursor.execute("SELECT * FROM books WHERE book_id = ? AND booktitle=? AND author=? AND year=? AND isbn=?",(str(BOOK_ID.get()),str(BOOKTITLE.get()),str(AUTHOR.get()),str(YEAR.get()),str(ISBN.get())))
            fetch=cursor.fetchall()
            for data in fetch:
                tree.insert('',"end",values=(data[1],data[2],data[3],data[4],data[5],data[6]))
            cursor.close()
            BOOK_ID.set("")
            BOOKTITLE.set("")
            AUTHOR.set("")
            STATUS.set("")
            ISBN.set("")
            YEAR.set("")
            txt_result.config(text="Successfully fetched data.",fg="black")
    
    
def Delete():
    tree.delete(*tree.get_children())
    if BOOK_ID.get()=="":
        txt_result.config(text="Please select the BOOK_ID to delete.",fg="red")
    else:
        Database()
        cursor.execute("SELECT * FROM books WHERE book_id=?",(str(BOOK_ID.get()),))
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert('',"end",values=(data[1],data[2],data[3],data[4],data[5],data[6]))
        cursor.close()
        BOOK_ID.set("")
        BOOKTITLE.set("")
        AUTHOR.set("")
        STATUS.set("")
        YEAR.set("")
        ISBN.set("")
    
    
def Exit():
    result = tkMessageBox.askquestion("Do you want to exit ? (y/n)",icon="warning")
    if result == 'yes':
        root.destroy()
        exit()
```

```python

BOOK_ID = StringVar()
BOOKTITLE = StringVar()
AUTHOR = StringVar()
STATUS = StringVar()
ISBN = StringVar()
YEAR = StringVar()
```

```python
Top = Frame(root, width=900, height=50, bd=8, relief='raise')
Top.pack(side=TOP)

Left = Frame(root,width=300,height=500,bd=8,relief='raise')
Left.pack(side=LEFT)

Right = Frame(root,width=600,height=500,bd=8,relief='raise')
Right.pack(side=RIGHT)

Forms = Frame(Left, width=300,height=450)
Forms.pack(side=TOP)

Buttons = Frame(Left,width=300,height=100,bd=8,relief='raise')
Buttons.pack(side=BOTTOM)

RadioGroup = Frame(Forms)
Available=Radiobutton(RadioGroup,text='Available',variable=STATUS,value='Available').pack(side=LEFT)
Issued=Radiobutton(RadioGroup,text='Issued',variable=STATUS,value='Issued').pack(side=LEFT)
```

```python
txt_title = Label(Top,width=900,font=('arial',24),text='Murugans: Library Management Application')
txt_title.pack()
txt_book_id = Label(Forms, text="Book_Id:", font=('arial',16), bd=15)
txt_book_id.grid(row=0, stick="e")
txt_booktitle = Label(Forms, text="Booktitle:", font=('arial',16), bd=15)
txt_booktitle.grid(row=1, stick="e")
txt_author = Label(Forms, text="Author:", font=('arial',16), bd=15)
txt_author.grid(row=2, stick="e")
txt_status = Label(Forms, text="Status:", font=('arial',16), bd=15)
txt_status.grid(row=3, stick="e")
txt_isbn = Label(Forms, text="ISBN:", font=('arial',16), bd=15)
txt_isbn.grid(row=4, stick="e")
txt_year = Label(Forms, text="Year:", font=('arial',16), bd=15)
txt_year.grid(row=5, stick="e")

txt_result = Label(Buttons)
txt_result.pack(side=TOP)

#=======================================================ENTRY WIDGET===========================================================#

book_id = Entry(Forms, textvariable=BOOK_ID, width=30)
book_id.grid(row=0, column=1)
booktitle = Entry(Forms, textvariable=BOOKTITLE, width=30)
booktitle.grid(row=1, column=1)
author = Entry(Forms, textvariable=AUTHOR, width=30)
author.grid(row=2, column=1)
RadioGroup.grid(row=3, column=1)
isbn = Entry(Forms, textvariable=ISBN, width=30)
isbn.grid(row=4, column=1)
year = Entry(Forms, textvariable=YEAR, width=30)
year.grid(row=5, column=1)

#=====================================================BUTTONS WIDGET===========================================================#

btn_Add = Button(Buttons, width=10, text="Add", command=Add)
btn_Add.pack(side=LEFT)
btn_Display = Button(Buttons, width=10, text="Display", command=Display)
btn_Display.pack(side=LEFT)
btn_Search = Button(Buttons, width=10, text="Search", command=Search)
btn_Search.pack(side=LEFT)
btn_Issue = Button(Buttons, width=10, text="Issue", command=Issue)
btn_Issue.pack(side=LEFT)
btn_Delete = Button(Buttons, width=10, text="Delete", state=NORMAL)
btn_Delete.pack(side=LEFT)
btn_exit = Button(Buttons, width=10, text="Exit", command=Exit)
btn_exit.pack(side=LEFT)


#=======================================================LISTS WIDGET===========================================================#

scrollbary = Scrollbar(Right, orient=VERTICAL)
scrollbarx = Scrollbar(Right, orient=HORIZONTAL)

tree = ttk.Treeview(Right, columns=("Book_id","Booktitle","Author","Status","ISBN","Year"))

scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT, fill=Y)

scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM, fill=X)

tree.heading('Book_id', text='Book_id', anchor=W)
tree.heading('Booktitle', text='Booktitle', anchor=W)
tree.heading('Author', text='Author', anchor=W)
tree.heading('Status', text='Status', anchor=W)
tree.heading('ISBN', text='ISBN', anchor=W)
tree.heading('Year', text='Year', anchor=W)

tree.column('#0', stretch=NO, minwidth=0, width=0)
tree.column('#1', stretch=NO, minwidth=0, width=80)
tree.column('#2', stretch=NO, minwidth=0, width=120)
tree.column('#3', stretch=NO, minwidth=0, width=80)
tree.column('#4', stretch=NO, minwidth=0, width=120)
tree.column('#5', stretch=NO, minwidth=0, width=120)

tree.pack()

#=====================================================INITIALIZATION===========================================================#

if __name__ == '__main__':
    root.mainloop()
```
